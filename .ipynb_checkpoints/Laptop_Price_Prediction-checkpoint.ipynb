```python
df['ram'] = (df['Ram'].str.extract(r'(\d+)').astype(int))
df['int'] = (df[0].str.extract(r'(\d+)').astype(int))
df['W'] = df['Weight'].str.extract(r'(\d+\.?\d*)').astype(float)
```

```python
df.head()
```

```python
df=df.drop(columns=['Unnamed: 0','Ram','Memory','Gpu',0,'Weight'])
```

```python
df.head()
```

```python
df = df[
    [
        'Company', 'TypeName', 'Inches', 'OpSys', 'W', 'fwe',
        'ram', 'int', 1, 4, 3,'efe','Price'
    ]
]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
df.rename(columns={
    "int": "CpuModel",
    1: "CpuParta",
    3: "CpuPartb",
    4: "CpuPartc"
}, inplace=True)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(df.iloc[:,0:12],df.iloc[:,-1],test_size=0.2)
```

```python
X_train.head()
```

```python
split_cols.columns = split_cols.columns.astype(str)
```

```python
X_train.head()
```

```python
cat_cols = ['Company', 'TypeName', 'OpSys', 'fwe','CpuParta','CpuPartb','CpuPartc','efe']
num_cols = ['Inches', 'ram', 'W', 'CpuModel']

trf1 = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols),
        ('num', 'passthrough', num_cols)
    ]
)
```

```python
trf2 = RandomForestRegressor(n_estimators=200,n_jobs=-1)
# trf2 = LinearRegression()
# trf2 = DecisionTreeRegressor()
# trf2 = BaggingRegressor(random_state=1)
```

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2)
])
```

```python
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred = pipe.predict(X_test)
scores = cross_val_score(pipe, X_test, Y_test, cv=10, scoring="r2")
print("Scores:", scores)
print("Mean Accuracy:", scores.mean())
```
